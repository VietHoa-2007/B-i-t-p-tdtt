W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của ke

```python
In = input().split()
dict1 = {}
for c in In:
    k, v = c.split(":")
    if k not in dict1.keys():
        dict1[k] = [v]
    else:
        dict1[k].append(v)
print(dict1)
```

W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...

```python
lis= list(map(int, input().split()))
dic={'positive':0, 'negatives':0, 'zeros':0}
for i in lis:
    if i>0:
        dic['positive']+= 1
    elif i <0:
        dic['negative']+= 1
    else:
        dic['zeros']+= 1
print(dic)
```
